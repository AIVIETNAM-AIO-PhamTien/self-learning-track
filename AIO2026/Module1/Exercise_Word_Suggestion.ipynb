## 1. Getting Max Over Kernel

Cho `num_list` và cửa sổ trượt kích thước `k` (k ≥ 1) di chuyển từ trái sang phải, trả về danh sách giá trị lớn nhất trong mỗi cửa sổ.

- Input: `num_list = [3, 4, 5, 1, -44, 5, 10, 12, 33, 1]`, `k = 3`
- Output: `[5, 5, 5, 5, 10, 12, 33, 33]`

```python
def max_kernel(num_list: list, k: int) -> list:
    """
    Trả về danh sách các giá trị lớn nhất trong mỗi cửa sổ con (window) kích thước k
    chạy trượt trên danh sách num_list.

    Parameters:
        num_list (list): Danh sách các số đầu vào.
        k (int): Kích thước của cửa sổ trượt.

    Returns:
        list: Danh sách các giá trị lớn nhất trong từng cửa sổ con liên tiếp.
    """
    result = []

    # Your Code Here
    # Gợi ý: trượt cửa sổ kích thước k qua num_list, mỗi bước lấy max của cửa sổ
    #        rồi thêm vào result (có thể dùng slicing num_list[i:i+k] và hàm max()).
    for i in range(0,len(num_list) - k + 1):
      result.append(max(num_list[i:i+k]))
    # End Code Here

    return result


# Kiểm tra hàm
assert max_kernel([3, 4, 5, 1, -44], 3) == [5, 5, 5]
num_list = [3, 4, 5, 1, -44, 5, 10, 12, 33, 1]
k = 3
print(max_kernel(num_list, k))
```

## 2. Character Counting

Đếm số lần xuất hiện của từng ký tự trong một từ, trả về dictionary `{ký_tự: số_lần}`. Chuỗi được chuyển về chữ thường trước khi đếm.

- Input: `"smiles"` → Output: `{"s": 2, "m": 1, "i": 1, "l": 1, "e": 1}`

```python
def count_character(word: str) -> dict:
    """
    Đếm số lần xuất hiện của từng ký tự trong chuỗi đầu vào.
    Chuỗi được chuyển về chữ thường trước khi đếm (vd: 'B' và 'b' xem như một).

    Parameters:
        word (str): Chuỗi ký tự cần thống kê.

    Returns:
        dict: Từ điển với mỗi ký tự là một khóa, giá trị là số lần xuất hiện.
    """
    character_statistic = {}

    # Your Code Here
    # Gợi ý: chuyển word về chữ thường, duyệt từng ký tự; nếu ký tự đã có trong
    #        character_statistic thì tăng value, ngược lại khởi tạo bằng 1.
    word = word.lower()
    for c in word:
      if c in character_statistic.keys():
        character_statistic[f'{c}'] += 1
      else:
        character_statistic[f'{c}'] = 1
    # End Code Here

    return character_statistic


# Kiểm tra hàm
assert count_character("Baby") == {'b': 2, 'a': 1, 'y': 1}
print(count_character('smiles'))
# {'s': 2, 'm': 1, 'i': 1, 'l': 1, 'e': 1}
```

## 3. Word Counting

Đọc nội dung file txt, đếm số lần xuất hiện của từng từ, trả về dictionary `{từ: số_lần}`. Tiền xử lý: chuyển chữ thường, bỏ dấu `.` và `,`, tách từ.

**Tải dữ liệu `P1_data.txt`** — trên Colab/Jupyter bỏ dấu `#` ở cell dưới; chạy local thì tải thủ công và đặt cạnh notebook.

```python
def preprocess_text(sentence: str) -> list[str]:
    """
    Tiền xử lý một câu: chuyển chữ thường, bỏ dấu chấm (.) và phẩy (,), tách thành danh sách từ.

    Parameters:
        sentence (str): Chuỗi câu đầu vào.

    Returns:
        list[str]: Danh sách các từ sau khi xử lý.
    """
    words = []

    # Your Code Here
    # Gợi ý: dùng .lower() để viết thường, .replace() để bỏ dấu '.' và ',',
    #        rồi .split() để tách câu thành danh sách từ.
    sentence = sentence.lower().replace(".","").replace(",","").split()
    for word in sentence:
      words.append(word)
    # End Code Here

    return words


# Kiểm tra hàm
sentence = 'I love AI. AI is not easy'
print(preprocess_text(sentence))
# ['i', 'love', 'ai', 'ai', 'is', 'not', 'easy']
```

```python
def count_word(file_path: str) -> dict[str, int]:
    """
    Đọc nội dung file txt và đếm số lần xuất hiện của từng từ sau khi tiền xử lý.

    Parameters:
        file_path (str): Đường dẫn đến file.

    Returns:
        dict[str, int]: Từ điển với key là từ (str), value là số lần xuất hiện (int).
    """
    counter = {}

    # Your Code Here
    # Gợi ý: mở file đọc toàn bộ nội dung, dùng preprocess_text để tách từ,
    #        rồi đếm số lần xuất hiện của mỗi từ vào counter.
    file = open(file_path, "r")
    content = file.read()
    file.close()

    preprocessed_text = preprocess_text(content)
    for word in preprocessed_text:
      if word in counter.keys():
        counter[f'{word}'] += 1
      else:
        counter[f"{word}"] = 1
    # End Code Here

    return counter


# Kiểm tra hàm (cần file ./P1_data.txt)
file_path = './P1_data.txt'
result = count_word(file_path)
assert result.get('who', 0) == 3
print(result.get('man', 0))
```

## 4. Levenshtein Distance

Tính khoảng cách chỉnh sửa tối thiểu giữa hai chuỗi (số phép chèn/xóa/thay thế ít nhất để biến `token1` thành `token2`), dùng quy hoạch động.

Phần khởi tạo ma trận và hàng/cột đầu tiên đã cho sẵn; bạn chỉ cần điền **quy tắc cập nhật** cho các ô còn lại.

- Ví dụ: `levenshtein_distance("kitten", "sitting")` → `3`

```python
def levenshtein_distance(token1: str, token2: str) -> int:
    """
    Tính khoảng cách Levenshtein (edit distance) giữa hai chuỗi bằng quy hoạch động.

    Parameters:
        token1 (str): Chuỗi nguồn.
        token2 (str): Chuỗi đích.

    Returns:
        int: Số phép chỉnh sửa tối thiểu (chèn, xóa, thay thế).
    """
    # Khởi tạo ma trận (len(token1)+1) x (len(token2)+1)
    distances = [[0] * (len(token2) + 1) for _ in range(len(token1) + 1)]

    # Hàng đầu và cột đầu: khoảng cách so với chuỗi rỗng
    for t1 in range(len(token1) + 1):
        distances[t1][0] = t1
    for t2 in range(len(token2) + 1):
        distances[0][t2] = t2

    # Điền các ô còn lại
    for t1 in range(1, len(token1) + 1):
        for t2 in range(1, len(token2) + 1):
            # Your Code Here
            # Gợi ý:
            # - Nếu token1[t1-1] == token2[t2-1]: lấy lại giá trị ô chéo distances[t1-1][t2-1].
            if token1[t1-1] == token2[t2-1]:
              distances[t1][t2] = distances[t1-1][t2-1]
            # - Ngược lại: lấy min của 3 ô (trái = insert, trên = delete, chéo = replace) rồi + 1.
            else:
              distances[t1][t2] = min(
                  distances[t1-1][t2],
                  distances[t1][t2-1],
                  distances[t1-1][t2-1]
              ) + 1

            # End Code Here

    return distances[len(token1)][len(token2)]


# Kiểm tra hàm
assert levenshtein_distance("hi", "hello") == 4
assert levenshtein_distance("kitten", "sitting") == 3
print(levenshtein_distance("hola", "hello"))
# 3
```
